```python
from pynq import Overlay, Interrupt
import time
import asyncio

# 1. Load the Overlay
ol = Overlay("hw_wrapper.bit")

# 2. Check IP Dictionary
print("Available IPs:", ol.ip_dict.keys())

# 3. Locate your hardware block name from the IP dictionary
calculator_axi = ol.calculator_wrapper_0
```

Test the registers

```python
# Let's test all the registers
# Readback the default values
a_in = calculator_axi.read(0x00)
b_in = calculator_axi.read(0x04)
c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
op = calculator_axi.read(0x10)
interrupt = calculator_axi.read(0x14)
print(f"a_in      = {hex(a_in)}")
print(f"b_in      = {hex(b_in)}")
print(f"c_out     = {hex(c_out)}")
print(f"rem_out   = {hex(rem_out)}")
print(f"op        = {hex(op)}")
print(f"interrupt = {hex(interrupt)}")
```

Now test a few values

```python
# Addition
print("Addition")
calculator_axi.write(0x00, 0x1)
calculator_axi.write(0x04, 0x1)
calculator_axi.write(0x10, 0x0)
# Readback result and interrupt
c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
interrupt = calculator_axi.read(0x14)
print(f"c_out     = {hex(c_out)}")
print(f"rem_out   = {hex(rem_out)}")
print(f"interrupt = {hex(interrupt)}")

# clear interrupt
calculator_axi.write(0x14, 0x1)
interrupt = calculator_axi.read(0x14)
print(f"interrupt = {hex(interrupt)}")

# Add 10 + 5
calculator_axi.write(0x00, 0xa)
calculator_axi.write(0x04, 0x5)
calculator_axi.write(0x10, 0x0)
# Readback result and interrupt
c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
interrupt = calculator_axi.read(0x14)
print(f"c_out     = {(c_out)}")
print(f"rem_out   = {(rem_out)}")
print(f"interrupt = {hex(interrupt)}")

# clear interrupt
calculator_axi.write(0x14, 0x1)
interrupt = calculator_axi.read(0x14)
print(f"interrupt = {hex(interrupt)}")

# Subtraction
print("subtraction")
calculator_axi.write(0x00, 0x1)
calculator_axi.write(0x04, 0x1)
calculator_axi.write(0x10, 0x1)
# Readback result and interrupt
c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
interrupt = calculator_axi.read(0x14)
print(f"c_out     = {(c_out)}")
print(f"rem_out   = {(rem_out)}")
print(f"interrupt = {hex(interrupt)}")

# clear interrupt
calculator_axi.write(0x14, 0x1)
interrupt = calculator_axi.read(0x14)
print(f"interrupt = {hex(interrupt)}")

# Add 10 + 5
calculator_axi.write(0x00, 0xa)
calculator_axi.write(0x04, 0x5)
calculator_axi.write(0x10, 0x1)
# Readback result and interrupt
c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
interrupt = calculator_axi.read(0x14)
print(f"c_out     = {(c_out)}")
print(f"rem_out   = {(rem_out)}")
print(f"interrupt = {hex(interrupt)}")

# clear interrupt
calculator_axi.write(0x14, 0x1)
interrupt = calculator_axi.read(0x14)
print(f"interrupt = {hex(interrupt)}")

# Multiplication
print("multiplication")
calculator_axi.write(0x00, 0x1)
calculator_axi.write(0x04, 0x1)
calculator_axi.write(0x10, 0x2)
# Readback result and interrupt
c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
interrupt = calculator_axi.read(0x14)
print(f"c_out     = {(c_out)}")
print(f"rem_out   = {(rem_out)}")
print(f"interrupt = {hex(interrupt)}")

# clear interrupt
calculator_axi.write(0x14, 0x1)
interrupt = calculator_axi.read(0x14)
print(f"interrupt = {hex(interrupt)}")

# Add 10 + 5
calculator_axi.write(0x00, 0xa)
calculator_axi.write(0x04, 0x5)
calculator_axi.write(0x10, 0x2)
# Readback result and interrupt
c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
interrupt = calculator_axi.read(0x14)
print(f"c_out     = {(c_out)}")
print(f"rem_out   = {(rem_out)}")
print(f"interrupt = {hex(interrupt)}")

# clear interrupt
calculator_axi.write(0x14, 0x1)
interrupt = calculator_axi.read(0x14)
print(f"interrupt = {hex(interrupt)}")

# Divider
print("division")
calculator_axi.write(0x00, 0x1)
calculator_axi.write(0x04, 0x2)
calculator_axi.write(0x10, 0x3)
# Readback result and interrupt
c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
interrupt = calculator_axi.read(0x14)

while interrupt == 0:
    interrupt = calculator_axi.read(0x14)
    print(f"interrupt = {hex(interrupt)}")

c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
print(f"c_out     = {(c_out)}")
print(f"rem_out   = {(rem_out)}")

# clear interrupt
calculator_axi.write(0x14, 0x1)
interrupt = calculator_axi.read(0x14)
print(f"interrupt = {hex(interrupt)}")

# Add 10 + 5
calculator_axi.write(0x00, 0xa)
calculator_axi.write(0x04, 0x5)
calculator_axi.write(0x10, 0x3)
# Readback result and interrupt
c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
interrupt = calculator_axi.read(0x14)

while interrupt == 0:
    interrupt = calculator_axi.read(0x14)
    print(f"interrupt = {hex(interrupt)}")

c_out = calculator_axi.read(0x08)
rem_out = calculator_axi.read(0x0C)
print(f"c_out     = {(c_out)}")
print(f"rem_out   = {(rem_out)}")

# clear interrupt
calculator_axi.write(0x14, 0x1)
interrupt = calculator_axi.read(0x14)
print(f"interrupt = {hex(interrupt)}")
```
